##EDA 2: Coin History

```python
#helper function to get historical prices, market cap and volume in usd for up to one year
def coin_USD_price_history(coin = "bitcoin", currency = "usd", days = 365): #enter coin id and resulting currency as strings
  coin_history_response = cg.get_coin_market_chart_by_id(coin, currency, 365)
  df_coin_history = pd.DataFrame.from_dict(coin_history_response)

  df_coin_history[["timestamp", f"{coin} price in USD"]] = pd.DataFrame(df_coin_history['prices'].tolist())
  df_coin_history[f"{coin} market_cap in USD"] = pd.DataFrame(df_coin_history['market_caps'].tolist())[1]
  df_coin_history[f"{coin} 24 total volume in USD"] = pd.DataFrame(df_coin_history['total_volumes'].tolist())[1]

  df_coin_history["timestamp"] = pd.to_datetime(df_coin_history["timestamp"], unit='ms')
  df_coin_history_final = df_coin_history.iloc[:,3:]

  return df_coin_history_final
```

##EDA 3: DEXES Volumes

```python
#helper function to get the historical volume for up to one year in BTC
def get_historical_dex_volume_in_BTC(exchange_id, time_window_in_days = 365): #enter exchange id as string

  json_volume_response = cg.get_exchanges_volume_chart_by_id(exchange_id, time_window_in_days)
  exchange_volume_df = pd.DataFrame(json_volume_response, columns = ["timestamp", f"exchange volume {exchange_id} in BTC"])
  exchange_volume_df["timestamp"] = pd.to_datetime(exchange_volume_df["timestamp"], unit='ms') #to convert unix timestamp in miliseconds to datetime

  return exchange_volume_df
```

```python
df_uniswap_v2 = get_historical_dex_volume_in_BTC("uniswap_v2", 365)
```

```python
df_uniswap_v3 = get_historical_dex_volume_in_BTC("uniswap_v3", 365)
```

```python
 #{'id': 'curve_ethereum', 'name': 'Curve (Ethereum)'}
df_curve = get_historical_dex_volume_in_BTC("curve_ethereum", 365)
```

```python
#helper function to (inner) merge the resulting dataframes for the different exchanges
def merge_df_on_timestamps(df1,df2, merge_method = "inner"):
  if (df1.iloc[:,0] == df2.iloc[:,0]).unique()[0] == True: #and (df2.iloc[:,0] == df3.iloc[:,0]).unique()[0] == True:
    merged_df = df1.merge(df2, on="timestamp", how = merge_method)#.merge(df3, on = "timestamp")
    return merged_df
  else:
    return "Timestamp do not match"
```

```python
df_merged1 = merge_df_on_timestamps( df_uniswap_v2,df_uniswap_v3)
df_merged2 = merge_df_on_timestamps(df_curve,df_merged1)
```

```python
# we can use the bitcoin price using the date (ignoring the effect that the hours differ, but on aggregate wouldn't be an issues)

#ignore the difference in hours
df_merged2["Date"] = df_merged2["timestamp"].dt.date
df_btc_price["Date"] = df_btc_price["timestamp"].dt.date

#merge on the same day
merged_dex_table_USD = df_merged2.merge(df_btc_price[["Date","bitcoin price in USD"]], on ="Date",  how = "left")

##calculate the 24 h volume in USD for the different exchanges

#transform all the values into floats to multiply later
merged_dex_table_USD['exchange volume curve_ethereum in BTC'] = merged_dex_table_USD['exchange volume curve_ethereum in BTC'].astype(float)
merged_dex_table_USD['exchange volume uniswap_v2 in BTC'] = merged_dex_table_USD['exchange volume uniswap_v2 in BTC'].astype(float)
merged_dex_table_USD['exchange volume uniswap_v3 in BTC'] = merged_dex_table_USD['exchange volume uniswap_v3 in BTC'].astype(float)
merged_dex_table_USD['bitcoin price in USD'] = merged_dex_table_USD['bitcoin price in USD'].astype(float)

#multiply using pandas inbuilt vectorization - thats why we dont a for loop
merged_dex_table_USD['exchange volume curve_ethereum in USD'] = merged_dex_table_USD['exchange volume curve_ethereum in BTC'] * merged_dex_table_USD['bitcoin price in USD']
merged_dex_table_USD['exchange volume uniswap_v2 in USD'] = merged_dex_table_USD['exchange volume uniswap_v2 in BTC'] * merged_dex_table_USD['bitcoin price in USD']
merged_dex_table_USD['exchange volume uniswap_v3 in USD'] = merged_dex_table_USD['exchange volume uniswap_v3 in BTC'] * merged_dex_table_USD['bitcoin price in USD']

pd.set_option('display.float_format', '{:.2f}'.format) #change from scientic notation to standard number
merged_dex_table_USD
```

```python
# Define the new order of the columns
new_column_order = ['timestamp', 'Date',
                    'exchange volume curve_ethereum in BTC', 'exchange volume curve_ethereum in USD',
                    'exchange volume uniswap_v2 in BTC', 'exchange volume uniswap_v2 in USD',
                    'exchange volume uniswap_v3 in BTC', 'exchange volume uniswap_v3 in USD',
                    'bitcoin price in USD']

# Reorder the columns
final_merged_dex_table_USD = merged_dex_table_USD[new_column_order]
final_merged_dex_table_USD
```

##EDA 4: Get Info via contract address

###Example of Uniswap version 2 (V2)
The following smart contracts are involved in this protocol:
* UniswapV2ERC20 — an extended ERC20 implementation that’s used for LP-tokens It additionally implements EIP-2612 to support off-chain approval of transfers.
* UniswapV2Factory — similarly to V1, this is a factory contract that creates pair contracts and serves as a registry for them. The registry uses create2 to generate pair addresses–we’ll see how it works in details.
* UniswapV2Pair — the main contract that’s responsible for the core logic. It’s worth noting that the factory allows to create only unique pairs to not dilute liquidity.
* UniswapV2Router — the main entry point for the Uniswap UI and other web and decentralized applications working on top of Uniswap.
* UniswapV2Library — a collection of helper functions that implement important calculations.

Source: https://medium.com/@bloqarl/uniswap-v2-complete-guide-to-understand-the-defi-protocol-d3a0e766cbdd#:~:text=Each%20Uniswap%20smart%20contract%2C%20or,the%20price%20of%20the%20assets.
